```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
plt.rcParams['figure.figsize']= [10,8]
```

```
flights=sns.load_dataset('flights')
#seaborn 에 내장된 데이터 프레임
flights.head()
```

```
df=flights.pivot('month','year','passengers')
df.head()
```

```
#heatmap을 그리는 데 matplotlib 보다 seaborn 이 간단하다.
plt.pcolor(df)

plt.title('Heatmap by plt.pcolor()',fontsize=20)

plt.xticks(np.arange(0.5,len(df.columns),1),df.columns)
plt.xlabel('Year',fontsize=14)

plt.yticks(np.arange(0.5,len(df.index),1),df.index)
plt.ylabel('Month',fontsize=14)

plt.colorbar() #오른 쪽 막대 바
plt.show()
```

```
ax=sns.heatmap(df)

plt.title("heatmap",fontsize=10)

plt.show()
```

```
# annot=True 일시 주석(annotate) 숫자들을 적어 넣음
# fmt='d' 는 정수 형태(integer format)로 숫자를 입력하라는 뜻입니다. 

sns.heatmap(df,annot=True,fmt='d')
plt.title('Annoteat cell with values')
plt.show()
```

```
sns.heatmap(df,cmap='RdYlGn_r',annot=True,fmt='d')
#cmap , colors map
```

```
# center the colormap at a specific value

sns.heatmap(df, center=df.loc['Jan', 1949])
plt.title('Center the colormap at Jan. 1949', fontsize=20)
plt.show()



#출처: https://rfriend.tistory.com/419 [R, Python 분석과 프로그래밍의 친구 (by R Friend)]
```
